```python
import re
```

```python
def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub("", text)
df = df.applymap(lambda x: remove_html_tags(x) if isinstance(x, str) else x)
```

```python
def remove_url(text):
    pattern=re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub("",text)
df = df.applymap(lambda x: remove_url(x) if isinstance(x, str) else x)
```

```python
import string
exclude=string.punctuation
def remove_punc1(text):
    return text.translate(str.maketrans("","",exclude))
df = df.applymap(lambda x: remove_punc1(x) if isinstance(x, str) else x)
```

```python
chat_words={
   " AFAIK":"As Far As I Know",
"AFK": "Away From Keyboard",
"ASAP":"As Soon As Possible",
"BTW":"By The Way",
"B4":"Before",
"LAMO":"Laugh My A.. Off",
"FYI":"For your information"
}
def chat_conversion(text):
    new_text=[]
    for w in text.split():
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

```python
df = df.applymap(lambda x: chat_conversion(x) if isinstance(x, str) else x)
```

```python
import textblob
from textblob import TextBlob
```

```python
def correct_spelling(text):
    blob = TextBlob(text)
    return str(blob.correct())
```

```python
df = df.applymap(lambda x:correct_spelling(x) if isinstance(x, str) else x)
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
def remove_stopwords(text):
    new_text=[]

    for word in text.split():
        if word in stopwords.words("english"):
            new_text.append("")
        else:
            new_text.append(word.strip())


    return " ".join(new_text).replace("   ","")
df = df.applymap(lambda x:remove_stopwords(x) if isinstance(x, str) else x)
```

```python
import spacy

# Load the English NLP model from spacy
nlp = spacy.load("en_core_web_sm")
```

```python
def tokenize_text(text):
    # Process the text with the spacy model
    doc = nlp(text)

    # Extract tokens from the processed document
    tokens = [token.text for token in doc]

    return tokens
```

```python
df = df.applymap(lambda x:tokenize_text(x) if isinstance(x, str) else x)
```

```python
from nltk.stem import PorterStemmer

def stemming(text):
    obj=PorterStemmer()

    stem_word=[obj.stem(word) for word in text.split()]

    return stem_word
```

```python
df = df.applymap(lambda x:stemming(x) if isinstance(x, str) else x)
```

```python
from nltk.stem import WordNetLemmatizer

def lammatization(text):
    words=text.split()

    lemmetizer=WordNetLemmatizer()

    lemetized_word=[lemmetizer.lemmatize(word) for word in words]

    return lemetized_word
```

```python
df = df.applymap(lambda x:lammatization(x) if isinstance(x, str) else x)
```
